## Ajuda on-line <a name="ajuda"></a> ([Início](#inicio))


---



Utilizando as funções de ajuda ```?``` e ```help()```

````r
?ls
````

````r
help(ls)
````

````r
help("ls")
````

As aspas são importantes para caracteres especiais

````r
?*
````

````r
help("*")
````

Por padrão, a função ```help()``` procura apenas por pacotes carregados na memória; mas também é possível buscar em um pacote específico ou em todos os pacotes com a opção ```try.all.packages = TRUE``` 

````r
help("bs")
````

````r
help("bs", package = "splines")
````

````r
help("bs", try.all.packages = TRUE)
````

Busca por uma palavra específica em todas as páginas de ajuda

````r
help.search("tree")
````

Para o caso de pacotes instalados recentementes, talvez seja necessário atualizar a base de dados utilizada

````r
help.search("tree", rebuild = TRUE)
````

Busca por todas as funções em memória cujos nomes contêm uma determinada string

````r
apropos("help")
````

R trabalha com objetos que são caracterizados por seus atributos, dois deles intrisecos são ```mode``` e ```length```

````r
x <- 1
mode(x)
length(x)
````

````r
A <- "Gomphotherium"; compar <- TRUE; y <- 2; z <- 1i
mode(A); mode(compar); mode(y); mode(z)
````

````r
N <- 2.1e23
N
````

````r
exp(x)
````

## Escrita de dados em um arquivo <a name="leitura-arquivo"></a> ([Início](#inicio))

---




Verificando o diretório corrente e a lista de objetos

````r
getwd()
````

````r
ls()
````

Salvando um objeto em um arquivo

````r
x; y
````

````r
write(x, file = "data.txt")
````

````r
write(y, file = "data.txt")
````

Salvando um objeto tabular (tipicamente um *data frame*) em um arquivo

````r
write.table(mycsvdata, file = "mycsvdata-local.csv")
````

Salvando um grupo de objetos, maneira mais fácil de transferir dados (*workspace* no jargão R) entre máquinas

````r
save(x, y, z, file = "xyz.RData")
````

Carregando dados salvos em um arquivo

````r
rm(x, y, z); ls()
````

````r
load("xyz.RData"); ls()
````

Salvando todo o ambiente (equivalente a ```save(list = ls(all = TRUE), file = ".RData")```)

````r
save.image()
````

````r
system("pwd")
````

## Gerando dados <a name="gerando-dados"></a> ([Início](#inicio))


---



### Sequências regulares



````r
x <- 1.5:30.5
x
````

O operador ":" tem prioridade sobre operadores aritméticos

````r
1:10-1
````

````r
1:(10-1)
````

A função **seq()** gera uma sequência de números da seguinte forma

````r
seq(from = 1, to = 5, by = 0.5)
````

Quando não indicado, assume-se os argumentos como início, fim e incremento da sequência

````r
seq(1, 5, 0.5)
````

Mas também pode ser da seguinte forma

````r
seq(length = 9, from = 1, to = 5)
````

Ou mesmo gerar a sequência manualmente, com a função **c()**

````r
c(1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5)
````

A função **rep()** cria um vetor de elementos iguais

````r
rep(1, 30)
````

````r
rep(seq(1,5,0.5), 2)
````

````r
rep("ifpb", 10)
````

A função **sequence()** cria uma série de sequências finalizadas pelos argumentos passados

````r
sequence(4:5)
````

````r
# não confundir com...
sequence(4,5)
````

````r
sequence(c(10, 5))
````

A função **gl()** gera níveis e é útil para gerar séries regulares de fatores

````r
# 3 níveis de fatores, 5 réplicas
gl(3, 5)
````

````r
gl(3, 5, length = 30)
````

````r
gl(2, 6, labels = c("Male", "Female"))
````

A função **expand.grid()** cria um *data frame* com todas as combinações de vetores ou fatores passados como argumentos

````r
expand.grid(altura=c(60, 80), largura=c(100, 300), sexo=c("Masculino", "Feminino"))
````

### Sequências aleatórias

| Distribuição de probabilidade | Função |
|---|---|
| Gaussian (normal) | rnorm(n, mean=0, sd=1) |
| exponential | rexp(n, rate=1) |
| gamma | rgamma(n, shape, scale=1) |
| Poisson | rpois(n, lambda) |
| Weibull | rweibull(n, shape, scale=1) |
| Cauchy | rcauchy(n, location=0, scale=1) |
| beta | rbeta(n, shape1, shape2) |
| ‘Student’ (t) | rt(n, df) |
| Fisher–Snedecor (F) | rf(n, df1, df2)|
| Pearson (χ2) | rchisq(n, df) | 
| binomial | rbinom(n, size, prob) |
| multinomial | rmultinom(n, size, prob) |
| geometric | rgeom(n, prob) |
| hypergeometric | rhyper(nn, m, n, k) |
| logistic | rlogis(n, location=0, scale=1) |
| lognormal | rlnorm(n, meanlog=0, sdlog=1) |
| negative binomial | rnbinom(n, size, prob) |
| uniform | runif(n, min=0, max=1) |
| Wilcoxon’s statistics | rwilcox(nn, m, n), rsignrank(nn, n) |

Formato geral é **rfunc(n, p1, p2...)**, onde **func** é a função de probabilidade, **n** é a quantidade de dados gerados e **p1, p2...** são os parâmetros da respectiva função

````r
rnorm(n = 10, mean = 0, sd = 1)
````

Similarmente, **dfunc()**, **pfunc()** e **qfunc()** são as funções que representam a densidade de probabilidade, a densidade cumulativa de probabilidade e o valor do quantil, respectivamente

````r
dnorm(10, mean = 0, sd = 1)
pnorm(10, mean = 0, sd = 1)
qnorm(0.025, mean = 0, sd = 1)
````
